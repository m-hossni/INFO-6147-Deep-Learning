```python
import matplotlib.pyplot as plt
import torchvision.transforms as transforms
import seaborn as sns
import json
```

```python
with open('../model_data/DL_project_models_tracking.json', 'r') as f:
    clf1 = json.load(f)

#with open('../model_data/DL_project_models_trackingclf2.1.json', 'r') as f:
#    clf2 = json.load(f)

with open('../model_data/DL_project_models_tracking_clf3.json', 'r') as f:
    clf3 = json.load(f)

with open('../model_data/DL_project_models_trackingclf4.json', 'r') as f:
    clf4 = json.load(f)

with open('../model_data/DL_project_models_trackingclf5.json', 'r') as f:
    clf5 = json.load(f)

with open('../model_data/DL_project_models_trackingclf4_2_y.json', 'r') as f:
    clf6 = json.load(f)

with open('../model_data/clf_7_logs.json', 'r') as f:
    clf7 = json.load(f)

with open('../model_data/resnet_18_logs.json', 'r') as f:
    resent = json.load(f)
```

```python
data_list= [clf1,clf3,clf4,clf5,clf6,clf7,resent]
```

```python
tr_losses = []
ts_losses = []
ts_accuracies = []
models = []
for clf in data_list:
    for key in clf.keys():
        tr_loss = clf[key]['train_losses']
        tr_losses.append(tr_loss)
        ts_loss = clf[key]['test_losses']
        ts_losses.append(ts_loss)
        ts_acc = clf[key]['test_accuracy']
        ts_accuracies.append(ts_acc)
        models.append(key)
        '''print('\n\nmodel {}\n\n'.format(key))
        plt.plot(tr_loss)
        plt.title("model: {} training losses".format(key))
        plt.xlabel("iterations")
        plt.ylabel("Losses")
        plt.show()

        plt.plot(ts_loss)
        plt.title("model: {} test losses".format(key))
        plt.xlabel("iterations")
        plt.ylabel("Losses")
        plt.show()

        plt.plot(ts_acc)
        plt.title("model: {} test accuracy".format(key))
        plt.xlabel("iterations")
        plt.ylabel("Accuracy")
        plt.show()'''
```

```python
colors = plt.get_cmap('Set1').colors
for i, dt in enumerate(tr_losses):
    plt.plot(dt, label=models[i], color=colors[i])
plt.title('Training loss curves for diffrenet models')
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.legend()
plt.show()

for i, dt in enumerate(ts_losses):
    plt.plot(dt, label=models[i], color=colors[i])
plt.title('Test loss curves for diffrenet models')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()


for i, dt in enumerate(ts_accuracies):
    plt.plot(dt, label=models[i], color=colors[i])

plt.plot(resnet_pretrained['resnet_model_18_pretrained']['test_accuracy'], label='resnet pre trained')

plt.title('Test accuracy curves for diffrenet models')
plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)")
plt.legend()
plt.show()
```
